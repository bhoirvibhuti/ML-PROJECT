**PIPELINE CREATION**

```python
h.select_dtypes(include=['object']).columns#get columns having string  datatype
```

```python
#dividing my dataset into x and y
X = h.drop("BUY/NOT", axis=1)
y = h["BUY/NOT"]
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
#pre-processing the numerical colums
from sklearn.pipeline import Pipeline, make_pipeline
numeric_preprocessor = Pipeline(
    steps=[
        ("imputation_mean", SimpleImputer( strategy="mean")),
        ("scaler", StandardScaler()),
    ]
)
```

```python
#pre-processing the categorical columns
categorical_preprocessor = Pipeline(
    steps=[
        (
            "imputation_constant",
            SimpleImputer( strategy="most_frequent"),
        ),
        ("onehot", OneHotEncoder(handle_unknown="ignore",sparse_output=False)),
    ]
)
```

```python
#transforming categorical columns to numeric ones so algos will accept it
preprocessor = ColumnTransformer(
    [
        ("categorical", categorical_preprocessor, ['Size (cm)', 'Weight (g)', 'Brix (Sweetness)', 'pH (Acidity)',
       'HarvestTime (days)', 'Ripeness (1-5)', 'Color', 'Variety',
       'Blemishes (Y/N)', 'Quality (1-5)']),
        ("numerical", numeric_preprocessor, ['Softness (1-5)']),
    ]
)
```

**APPLYING LOGISTIC REGRESSION MODEL**

```python
from sklearn.linear_model import LogisticRegression
pipe = make_pipeline(preprocessor, LogisticRegression(max_iter=500))
pipe
```

**CALCUATING MY MODEL ACCURACY**

```python
pipe.fit(X_train,y_train)
predictions = pipe.predict(X_test)
print(predictions)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```
